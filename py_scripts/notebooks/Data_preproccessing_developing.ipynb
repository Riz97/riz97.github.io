```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_csv("../../data/Life_Expectancy_Data.csv")
dataframe['Country'] = dataframe['Country'].replace('Russian Federation', 'Russia')
dataframe['Country'] = dataframe['Country'].replace('Viet Nam', 'Vietnam')
dataframe['Country'] = dataframe['Country'].replace('Venezuela (Bolivarian Republic of)', 'Venezuela')
```

```python
dataframe.head()
```

```python
dataframe.dtypes
```

```python
dataframe1=dataframe.rename(columns={"Life expectancy ":"Life_expectancy"})
```

```python
#keep only the last year and Developing
dataframe1=dataframe1[dataframe1["Year"]==2015]
dataframe1=dataframe1[dataframe1["Status"]=="Developing"]

developing=dataframe.rename(columns={"Life expectancy ":"Life_expectancy"})
developing_countries = developing[developing["Status"]=="Developing"]
```

```python
dataframe1
```

```python
dataframe1=dataframe1[["Country","Life_expectancy"]]
```

```python
dataframe1.to_csv("../../data/developing/developing-1.csv", sep=",", index = False)
```

```python
dataframe2=dataframe.rename(columns={"Life expectancy ":"Life_expectancy", "Hepatitis B":"HepatitisB", "Diphtheria ":"Diphtheria"})

#keep only the last year and Developing
dataframe2=dataframe2[dataframe2["Year"]==2015]
dataframe2=dataframe2[dataframe2["Status"]=="Developing"]
dataframe2=dataframe2[["Country","HepatitisB", "Polio", "Diphtheria","Population"]]
dataframe2["Sum"]=dataframe2["HepatitisB"]+dataframe2["Polio"]+dataframe2["Diphtheria"]
dataframe2.sort_values(by=["Population","Sum"], 
                         ascending=False, 
                         inplace=True)
dataframe2=dataframe2[["Country","Population", "HepatitisB","Polio","Diphtheria"]]
dataframe2=dataframe2.head(10)
dataframe2
```

```python
dataframe2.to_csv("../../data/developing/developing-2.csv", sep=",", index = False)
```

```python
dataframe3=dataframe.rename(columns={"Life expectancy ":"Life_expectancy", "Adult Mortality":"Adult_deaths","under-five deaths ":"Under_five_deaths", "infant deaths":"Infant_deaths"})
dataframe3=dataframe3[dataframe3["Year"]==2015]
dataframe3=dataframe3[dataframe3["Status"]=="Developing"]
dataframe3=dataframe3[["Country","Life_expectancy","Adult_deaths", "Under_five_deaths", "Infant_deaths", "Population"]]
dataframe3=dataframe3.dropna()
dataframe3.sort_values(by="Life_expectancy", 
                         ascending=True, 
                         inplace=True)
dataframe3 = dataframe3.astype({'Adult_deaths':'int',"Population":"int"})

dataframe3=dataframe3.head(10)
dataframe3
```

```python
developing_countries_life = developing_countries[["Country" , "Year", "Life_expectancy" ]]
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "San Marino"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Dominica"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Monaco"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Niue"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Palau"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Saint Kitts and Nevis"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Tuvalu"].index)
developing_countries_life = developing_countries_life.drop(developing_countries_life[developing_countries_life["Country"] == "Marshall Islands"].index)

developing_countries_life
```
