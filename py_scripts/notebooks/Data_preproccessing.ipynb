```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_csv("../../data/Life_Expectancy_Data.csv")
dataframe=dataframe.rename(columns={"Life expectancy ":"Life_expectancy"})
```

```python
dataframe.head()
```

```python
developed_countries = dataframe.loc[dataframe["Status"] == "Developed"].reset_index(drop = True)

developed_countries.head()
```

```python
developed_countries_life = developed_countries[["Country" , "Year", "Life_expectancy" ]]

developed_countries_life
```

```python
developed_countries_life_15 = developed_countries_life[developed_countries_life["Year"] == 2015].reset_index(drop = True).drop("Year" , axis = 1)

developed_countries_life_15
```

```python
developed_countries_life_15.to_csv("../../data/developed1-2.csv", sep=",", index = False)
```
