```python
df4=dataframe[["Country","Year","Status","Population","Life_expectancy"]]
df4 = df4[df4['Year'] ==2015]
df4 = df4.dropna()
df4.sort_values(by="Life_expectancy", 
                         ascending=False, 
                         inplace=True)
df4["Group"]=pd.cut(df4["Life_expectancy"], bins=[50,55,60,65,70,75,80,85,90], labels=["A","B","C","D","E","F","G","H"])

df4["Population"] = df4["Population"].replace([29491.0] , 2905000) 
df4["Population"] = df4["Population"].replace([984328.0] , 9000000) 
df4["Population"] = df4["Population"].replace([42364.0] , 4236400)
df4["Population"] = df4["Population"].replace([135876.0	] , 10360000)
df4["Population"] = df4["Population"].replace([518867.0] ,5188670)
df4["Population"] = df4["Population"].replace([56964.0] , 569640)
df4["Population"] = df4["Population"].replace([263531.0	] ,2635310	)
df4["Population"] = df4["Population"].replace([116985.0] , 1169850)
df4["Population"] = df4["Population"].replace([127141.0] , 127141000)
df4["Population"] = df4["Population"].replace([673582.0] , 67358200)
df4["Population"] = df4["Population"].replace([33815.0] , 338150)
df4=df4[["Country","Status","Population","Life_expectancy","Group"]]
df4.to_csv("../../data/comparison/comparison-4.csv", sep=",", index = False)
```
